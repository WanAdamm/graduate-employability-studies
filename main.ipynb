```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("ge degree.csv")
```

```python
df.head(),df.nunique()
```

```python
ge_by_uni = df.groupby('Institutions')["GE (%)"]
```

```python
stat = ge_by_uni.describe()
stat.head()
```

```python
stat.to_csv('uni employability all.csv')
```

```python
stat_df = pd.read_csv("uni employability all.csv")
stat_df = stat_df.sort_values(by=['mean'], ascending=True)
```

```python
ax = sns.barplot(x= stat_df['Institutions'], y = stat_df['mean'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.ylabel("graduate employability mean")
plt.show()
```
